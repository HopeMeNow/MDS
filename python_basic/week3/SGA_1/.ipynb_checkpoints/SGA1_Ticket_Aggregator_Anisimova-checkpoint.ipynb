```python
import re
import random
from typing import Type, TypeVar, Dict, Callable, List


def get_number(message: str) -> float or '':
    answer = input(message)
    # If user wants to exit, just pass this information onward
    if answer == 'e' or answer == 'E':
        return answer.lower()
    num: List[str] = answer.split('.')
    # We always have at least one element. It can be empty
    parsed_num = ''.join(re.findall('\d+', num[0]))
    # But if we have more than one element, so there is a float point. Let's use it to create float number.
    # It can be more than one dot in the string, but it doesn't make a sence for numbers, so let's just ignore them
    if len(num) > 1:
        parsed_num = parsed_num + '.' + ''.join(re.findall('\d+', num[1]))
    # if we don't have any digits in the input string, parsedNum is empty string, so let's return it as it
    # else - parsedNum is a valid number in string format, so let's convert it to float
    return parsed_num if parsed_num == '' else abs(float(parsed_num))

def get_bool(message: str) -> bool or None:
    answer: str = input(message + ' Type "y" if Yes, "n" if No \n')
    # If user wants to exit, just pass this information onward
    if answer == 'e' or answer == 'E':
        return answer.lower()
    # Decipher convention to boolean
    elif answer == 'y' or answer == 'Y':
        return True
    elif answer == 'n' or answer == 'N':
        return False
    else:
    # If user enter wrong value return None and handle it later
        return None
```

```python
# Create decorators and decorate common getters to pass custom messages and do custom conversion

def get_price_decorator(get_func: Type[get_number]) -> Callable[[str], float]:
    def get_price_func(message: str = '- What is the price of ticket in dollars?\n') -> float or '':
        return get_func(message)
    return get_price_func


def get_price_confirmation_decorator(get_func: Type[get_bool]) -> Callable[[float, str], float]:
    def get_price_confirmation_func(value: float, message: str):
        if message == '':
            return get_func(f'- Is this correct price: ${value}?')
        else:
            return get_func(message)
    return get_price_confirmation_func


get_price = get_price_decorator(get_number)
get_price_confirmation = get_price_confirmation_decorator(get_bool)

def get_n_transfers_decorator(get_func: Type[get_number]) -> Callable[[str], int]:
    def get_n_transfers_func(message: str = '- What is the number of transfers?\n') -> int or '':
        num: float or '' = get_func(message)
        return num if num == '' or num == 'e' else abs(int(num))
    return get_n_transfers_func


def get_n_transfers_confirmation_decorator(get_func: Type[get_bool]) -> Callable[[int, str], int]:
    def get_n_transfers_confirmation_func(value: int, message: str) -> int:
        if message == '':
            return get_func(f'- Is this correct number of transfers: {value}?')
        else:
            return get_func(message)
    return get_n_transfers_confirmation_func


get_n_transfers = get_n_transfers_decorator(get_number)
get_n_transfers_confirmation = get_n_transfers_confirmation_decorator(get_bool)
```

```python
def get_boolean_criterion(message: str) -> bool:
    answer = get_bool(message)
    # If user wants to exit, just pass this information onward
    if answer == 'e':
        return answer
    # If answer in right format, return it
    elif answer == True or answer == False:
        return answer
    # If user enter wrong value, call yourself until user enter correct value
    else:
        return get_boolean_criterion('- You entered incorrect symbol.')
```

```python
def get_numeric_criterion(value_getter: Type[get_number], confirmation_getter: Type[get_bool]) -> float or int:
    def get_value() -> float or int:
        value = value_getter()
        i = 0
        if value == 'e':
            return value
        if value == '':
            # Request value until user enter correct value
            while value == '':
                value = value_getter(messages[i])
                # Enumerate funny messages
                if i == len(messages) - 1:
                    i = 0
                else:
                    i += 1
        return value

    def get_confirmation(value: float, message: str = '') -> float or int:
        # Since we clear the value and compose the final result at its discretion, let's clarify
        # if we understand user right way
        answer = confirmation_getter(str(value), message)
        if answer == 'e':
            return answer
        if answer == True:
            return value
        elif answer == False:
            value = get_value()
            return get_confirmation(value)
        else:
            return get_confirmation(value, '- You entered incorrect symbol.')
        
    value = get_value()
    if value == 'e':
        return value
    return get_confirmation(value)
```

```python
def boolAnswerToStr(boolAnswer: bool) -> str:
    if boolAnswer:
        return 'Yes'
    else:
        return 'No'

def analizer(price: float, n_transfers: int, refund: bool, luggage: bool) -> str:
    if price < 200 and n_transfers < 2 and refund and luggage:
        return 'The best'
    elif price >= 200 and price <= 250 and n_transfers < 3:
        return 'Good enough'
    elif price > 250 and n_transfers > 2:
        return 'Bad'
    else:
        return 'Other'
```

```python
def interface():        
    # Since we forced to check after each criteria does user want to exit, it
    # is more convenient to request information in loop, so let's create the array for it
    getters_array = [
        {
            'name': 'price',
            'get_value': get_numeric_criterion,
            'get_value_args': [get_price, get_price_confirmation]
        },
        {
            'name': 'n_transfers',
            'get_value': get_numeric_criterion,
            'get_value_args': [get_n_transfers, get_n_transfers_confirmation]
        },
        {
            'name': 'refund',
            'get_value': get_boolean_criterion,
            'get_value_args': ['Is there opportunity to refund the ticket?']
        },
        {
            'name': 'luggage',
            'get_value': get_boolean_criterion,
            'get_value_args': ['Is luggage included in the ticket cost?']
        },
    ]
    
    values = {}
    
    for item in getters_array:
        value = item['get_value'](*item['get_value_args'])
        if value == 'e':
            return
        values[item['name']] = value
    
    category = analizer(**values)
    
    
    print(
        '\n-------------------------------------------\n' +
        'OFFER FEATURES' +
        '\n-------------------------------------------\n' +
        '{:30s}  {:7.2f}\n'.format('Price: ', values['price']) +
        '{:30s}  {:7d}\n'.format('Number of transfers: ', values['n_transfers']) +
        '{:34s}  {:7s}\n'.format('Opportunity to refund: ', boolAnswerToStr(values['refund'])) +
        '{:34s}  {:7s}\n'.format('Luggage is included: ', boolAnswerToStr(values['luggage'])) +
        '\n-------------------------------------------\n' +
        '{:32s}  {:7s}\n'.format('Offer category: ', category)
    )
```
