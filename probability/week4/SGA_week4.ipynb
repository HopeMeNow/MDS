```python
import numpy as np
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
%matplotlib inline

n = 2**10

x = np.linspace(0, 10, n) 
a = -100
loc = 9
var = 3

mean = skewnorm.mean(a, loc=loc, scale=var)
median = skewnorm.median(a, loc=loc, scale=var)

plt.axvline(mean, color='red', linewidth=0.6, label='mean')
plt.axvline(median, color='green', linewidth=0.6, label='median')

print("\033[91m {}\033[00m {}" .format('mean', mean))
print("\033[92m {}\033[00m {}".format('median', median))

plt.plot(x, skewnorm.pdf(x, a, loc=loc, scale=var))

plt.show();
```
