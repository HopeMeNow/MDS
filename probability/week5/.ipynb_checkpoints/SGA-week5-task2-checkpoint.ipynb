```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import math

import sympy as sy
```

```python
x, y = np.mgrid[-4:4:100j, -4:4:100j]
z = (1./(2*math.pi))*math.e**(-(x**2+y**2)/2.)

fig, ax = plt.subplots()

ax.contour(z)

fig.set_figwidth(8)    #  ширина и
fig.set_figheight(8)    #  высота "Figure"

plt.show()
```

```python
delta = 0.025
x = np.arange(-2, 2, delta)
y = np.arange(-2, 2, delta)

z = np.array([])

for i in range(len(x)):
    temp = np.array([])
    for j in y:
        temp = np.append(temp, math.e**(-(x[i]**2+j**2)/2.))
    if len(z) == 0:
        z = np.append(z, temp)
    else:
        z = np.vstack((z, temp))

cs = plt.contourf(z, levels=[0.05, 0.1, 0.15])
cs.changed()
```
